```python
#언론사이름 별 언론사가 한해동안 몇건의 기사썼는지  엑셀로 만들면됨. 

from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
driver =webdriver.Chrome('C:/chromedriver')
```

```python
#네이버 옵션에서 뉴스 메뉴에 기간설정까지 해놓은 url
driver.get('https://search.naver.com/search.naver?where=news&query=sk%EC%9D%B4%EB%85%B8%EB%B2%A0%EC%9D%B4%EC%85%98&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds=2019.01.01&de=2019.12.31&docid=&nso=so%3Ar%2Cp%3Afrom20190101to20191231%2Ca%3Aall&mynews=0&refresh_start=0&related=0')
```

```python
#언론사 버튼 클릭
# driver.find_element_by_xpath("""//*[@id="news_popup"]/a""").click()
```

```python
# 언론사 이름 긁어오기
# label 태그에 언론사 이름이 들어가 있어서 전부 긁어옴

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
viewlst = soup.find_all('label')
```

```python
viewlst = [i.get_text() for i in viewlst]
len(viewlst)
```

```python
# 근데 리스트에 앞뒤로 언론사 리스트 외의 목록도 딸려와서 따로 정리작업 했음
agency_result=[]
for i in range(51,1882):
    agency_result.append(viewlst[i])
len(agency_result)
```

```python
# 중복 제거
a = set(agency_result)
agency_result = list(a)
len(agency_result)
```

```python
# 언론사 타이틀도 섞여있어서 제거 작업 진행
viewtit = soup.find_all('div', class_ = 'viewtit')
```

```python
viewtit = [i.get_text() for i in viewtit]
```

```python
# 공백 제거
viewtit2 = []
for i in range(0,8):
    viewtit2.append(viewtit[i].strip())
viewtit2
```

```python
# 일간지는 위에서 미리 빼놓아서 1번째 인덱스부터 제거
for i in range(1,8):
    j = agency_result.index(viewtit2[i])
    agency_result.pop(j)
len(agency_result)
```

```python
result = {}

for i in agency_result:
    count=0
    driver =webdriver.Chrome('C:/chromedriver')
    driver.get('https://search.naver.com/search.naver?where=news&query=sk%EC%9D%B4%EB%85%B8%EB%B2%A0%EC%9D%B4%EC%85%98&sm=tab_opt&sort=0&photo=0&field=0&reporter_article=&pd=3&ds=2019.01.01&de=2019.12.31&docid=&nso=so%3Ar%2Cp%3Afrom20190101to20191231%2Ca%3Aall&mynews=0&refresh_start=0&related=0')
    time.sleep(1)
    driver.find_element_by_xpath("""//*[@id="news_popup"]/a""").click() #언론사 버튼 클릭
    driver.find_element_by_xpath('//*[@title="{}"]'.format(i)).click() #언론사 하나 선택
    driver.find_element_by_xpath("""//*[@id="_nx_option_media"]/div[2]/div[3]/button[1]/span""").click() #확인버튼
    time.sleep(1)
    try:
        count = driver.find_element_by_xpath("""//*[@id="main_pack"]/div[2]/div[1]/div[1]/span""").text  #언론사별 건수
    except:
        pass
    result.update({i:count})
    time.sleep(1)
    driver.close()
    time.sleep(1)


# 기사 없는 언론사는 오류 뜬다..
```
